# Exploratory Data Analysis  
Since the main goal of analysis is to decide which type of transmission is better for obtaining
higher mpg, let us check the relation between mpg and am.

```python
sns.set()
plt.rcParams['figure.figsize']=(10,5)
ax = sns.boxplot(x="am", y="mpg", data=df1)
plt.title("Relation between mpg and am")
plt.ylabel('Miles Per Gallon')
plt.xlabel('Transmission (0=Automatic,1=Manual');
```

```python
sns.lmplot(x="hp",y="mpg",hue="vs", data=df1)
```

# Inferences:
It conveys that for staright engines, lesser horsepower yields higher mpg
For V-shaped engines, inspite of the higher horse power the values for mpg has fallen

```python
x = df1.drop(['model','drat','qsec','gear','carb','maker'],axis = 1)
fig , ax = plt.subplots(figsize = (8,8))
sns.heatmap(x.corr(),annot=True,center = 0 , cmap ='PuRd_r')
plt.show()
```

5) VS and AM are Positivly correlating with MPG

# Hypothesis Test and Confidence Interval

Ho : The null hypothesis states that there is no difference between the transmission types.  
Ha : The alternative hypothesis states that there is a significant difference between the mpg for
different transmission types.  
T-Test transmission type and MPG

```python
corr, pvalue = scipy.stats.pearsonr(df1['mpg'],df1['am'])
print('correlation coefficient: ', corr)
print('P-value: ' , pvalue)
```

```python
sns.set_style("whitegrid")
plt.rcParams['figure.figsize']=(5,5)
ax=sns.barplot(x="am", y="mpg", data=df1)
plt.title("Relation between mpg and am")
plt.ylabel('Miles Per Gallon')
plt.xlabel('Transmission (0=Automatic,1=Manual')
```

# **Conclusion:**




There is difference in MPG based on transmission type.
